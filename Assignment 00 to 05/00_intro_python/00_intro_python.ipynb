```python
# This program adds two numbers provided by the user.

print("01_add_two_numbers")
def main():
    num1 = int(input("Enter the first number: "))
    num2 = int(input("Enter the second number: "))
    total_sum = num1 + num2
    print(f"The sum of {num1} and {num2} is: {total_sum}")

if __name__ == "__main__":
    main()
```

```python
# This Program converts fahrenheit to celsius
print("03_fahrenheit_to_celsius")
def fahrenheit_to_celsius(fahrenheit):
    return (fahrenheit - 32) * 5.0 / 9.0

def main():
    fahrenheit = float(input('Enter temperature in Fahrenheit: '))
    celsius = fahrenheit_to_celsius(fahrenheit)
    print(f'Temperature: {fahrenheit}F = {celsius}C')

if __name__ == '__main__':
    main()
```

```python
# This program tells how old each person is:
print("04_how_old_are_they")
def main():
    anton_age = 21
    beth_age = anton_age + 6
    chen_age = beth_age + 20
    drew_age = chen_age + anton_age
    ethan_age = chen_age

    print(f'Anton is {anton_age}')
    print(f'Beth is {beth_age}')
    print(f'Chen is {chen_age}')
    print(f'Drew is {drew_age}')
    print(f'Ethan is {ethan_age}')

if __name__ == '__main__':
    main()
```
