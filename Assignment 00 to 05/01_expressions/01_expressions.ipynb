```python
import random
print("01_dicesimulator")

num_sides: int =6
def roll_dice():
    dice1:int = random.randint(1,num_sides)
    dice2:int = random.randint(1,num_sides)
    total:int = dice1 + dice2
    print(f"You rolled a {dice1} and a {dice2} for a total of {total}")
    return total

def main():
    user_input:int = int(input("Guess the number between 2 and 12: "))
    total1 = roll_dice()
    total2 = roll_dice()

    if total1 == user_input or total2 == user_input:
        print("You win!")
    else:
        print("You lose!")

# Python file to call the main() function.
if __name__ == '__main__':
    main()
```

```python
print("02_e=mc2")
C: int = 299792458 # The speed of light in m/s
def main():
    mass_input: float = float(input("Enter the mass: "))
    energy_in_joules: float = mass_input * C ** 2
    print ("e = m * C^2")
    print(f"m = {mass_input} kg")
    print(f"c = {C} m/s")
    print(f"joules_of_energy = {energy_in_joules} J")

if __name__ == '__main__':
    main()
```

```python
print("03_feet_to_inches")
inches_in_feet = 12
def main():
    feet = float(input("Enter the number of feet: "))
    conversion = feet * inches_in_feet
    print(f"{feet} feet is equal to {conversion} inches.")

if __name__ == '__main__':
    main()
```

```python
import math
print("04_pythagorean_theorem")
def main():
    ab: float = float(input("Enter the length of the first side AB: "))
    ac: float = float(input("Enter the length of the second side AC: "))
    hypotenuse_bc: float = math.sqrt(ab**2 + ac**2)
    print(f"The length of the hypotenuse is BC: {hypotenuse_bc:.2f}")

if __name__ == "__main__":
    main()
```

```python
print("05_remainder_division")
def main():

    dividend: int = int(input("Enter the first number is divided: "))
    divisor: int = int(input("Enter the second number is divide by: "))
    quotient: int = dividend // divisor
    remainder: int = dividend % divisor

    print(f"The result of this division is {quotient} with a remainder of {remainder}")
if __name__ == '__main__':
    main()
```

```python
print("06_seconds_in_year")

days_per_year : int = 365
hours_per_day : int = 24
minutes_per_hour : int = 60
seconds_per_minute : int = 60

def main():
    seconds_in_year : int = days_per_year * hours_per_day * minutes_per_hour * seconds_per_minute
    print(f"There are {seconds_in_year} seconds in a year")
if __name__ == '__main__':
    main()
```
